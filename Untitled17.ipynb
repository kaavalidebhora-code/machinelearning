```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Temperature":[25,30,28,27,35,22],
    "Humidity":[60,65,70,68,75,80],
    "WindSpeed":[10,12,11,14,13,9],
    "Precipitation":[2,5,0,1,2,3],
    "Pressure":[1012,1015,1010,1011,1020,1008],
    "WeatherCondition":[0,1,2,0,2,2]
}

df = pd.DataFrame(data)

print("Dataset")
print(df)
```

```python
X = df.drop("WeatherCondition", axis=1)
y = df["WeatherCondition"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print("Scaled Features:")
print(X_scaled)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

filter_method = SelectKBest(score_func=f_classif, k=3)

X_filter = filter_method.fit_transform(X_scaled, y)

print("Selected Features (Filter Method):")
print(X_filter)
```

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

rfe = RFE(model, n_features_to_select=3)

X_wrapper = rfe.fit_transform(X_scaled, y)

print("Selected Features (Wrapper Method):")
print(X_wrapper)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1)

lasso.fit(X_scaled, y)

print("Feature Importance (Embedded Method):")
print(lasso.coef_)
```
