```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\HP\Downloads\weather_dataset.csv")

print("Original Dataset:")
print(df)
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())

print("After filling numeric missing values:")
print(df)
```

```python
df['WeatherCondition'] = df['WeatherCondition'].fillna(df['WeatherCondition'].mode()[0])

print("After filling categorical missing values:")
print(df)
```

```python
Q1 = df['Temperature'].quantile(0.25)
Q3 = df['Temperature'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Temperature'] < lower) | (df['Temperature'] > upper)]

print("Detected Outliers:")
print(outliers)
```

```python
df['Temperature'] = np.where(df['Temperature'] > upper, upper, df['Temperature'])
df['Temperature'] = np.where(df['Temperature'] < lower, lower, df['Temperature'])

print("Dataset after treating outliers:")
print(df)
```

```python
print("Final Preprocessed Dataset:")
print(df)
```
